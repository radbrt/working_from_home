# This is simply reading and restructuring Mechanical Turk output

```python
import pandas as pd
import json
```

```python
mt = open('data/mechanicalturk_annotations.json', 'r').readlines()
```

```python
len(mt)
```

This is a probably correct number of occupations. I haven't checked in detail with how many occupations are actually in the ISCO-08 standard, but this is in the ballpark. Parsing the descriptions file was tedious...

```python
mt_jsons = [json.loads(item) for item in mt]
```

```python
mt_jsons[0]
```

```python
mt_data = [{'ISCO': item['isco'], 
            'wfh': item['simpleisco-clone-metadata']['class-name'], 
            'confidence': item['simpleisco-clone-metadata']['confidence']} for item in mt_jsons]
```

```python
df = pd.DataFrame(mt_data)
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['wfh'].value_counts()
```

```python
df['ISCO1'] = df['ISCO'].apply(lambda x: x[0])
df['wfh_dummy'] = df['wfh'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
df['wfh_dummy'].mean()
```

Based on the mechanical turk data,  30% of all occupations (not employees) are eligeble for working remotely. Due to imprecisions in register data, estimating the true number of remote jobs will involve some uncertainty in addition to the uncertainty of the annotations.

```python
df.groupby('ISCO1').agg({'wfh_dummy': ['mean']})
```
